````python
#comprobar la distribucion de las variables cuantitativas 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import normaltest

#variables cuantitativas
cuantitativas= [
    "Id","MSSubClass","LotFrontage","LotArea","OverallQual","OverallCond","YearBuilt","YearRemodAdd","MasVnrArea","BsmtFinSF1",
    "BsmtFinSF2","BsmtUnfSF","TotalBsmtSF","1stFlrSF","2ndFlrSF","LowQualFinSF","GrLivArea","BsmtFullBath","BsmtHalfBath",
    "FullBath","HalfBath","BedroomAbvGr","KitchenAbvGr","TotRmsAbvGrd","Fireplaces","GarageYrBlt","GarageCars","GarageArea","WoodDeckSF","OpenPorchSF","EnclosedPorch","3SsnPorch","ScreenPorch","PoolArea","MiscVal","MoSold","YrSold"
]
#para el primer dataframe
# Convertir columnas a valores numéricos, forzando errores a NaN (not a number), esto para evitar problemas con el tipo de dato para analizar. 
for col in cuantitativas:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')

# Función para revisar la normalidad
def revisar_normalidad(data):
    stat, p_value = normaltest(data)
    if p_value > 0.05:
        return "Distribución Normal"
    else:
        return "No Distribución Normal"
    
# Graficar histogramas y revisar normalidad
for var in cuantitativas:
    plt.figure(figsize=(10, 6))
    sns.histplot(df1[var].dropna(), kde=True, bins=20)  # Eliminar NaN antes de graficar
    plt.title(f"Distribución de {var}")
    plt.show()
    
    resultado_normalidad = revisar_normalidad(df1[var].dropna())  # Eliminar NaN antes de la prueba
    print(f"Resultado de normalidad para {var}: {resultado_normalidad}")
````

Como podemos observar en cada tabla, no todas las variables cuantitativas tienen una distribución normal, lo que sugiere una gran variabilidad en los datos.

````python
#variables cualitativas

#seleccionar las variables cualitativas
cualitativas=[
    "MSZoning","Street","Alley","LotShape","LandContour","Utilities","LotConfig","LandSlope","Neighborhood","Condition1","Condition2","BldgType",
    "HouseStyle","RoofStyle","RoofMatl","Exterior1st","Exterior2nd","MasVnrType","ExterQual","ExterCond",
    "Foundation","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","Heating",
    "HeatingQC","CentralAir","Electrical","KitchenQual","Functional","FireplaceQu","GarageType","GarageFinish",
    "GarageQual","GarageCond","PavedDrive","PoolQC","Fence","MiscFeature","SaleType","SaleCondition"
]

# Crear tabla de frecuencias para cada variable cualitativa
for col in cualitativas:
    print(f"\n🔹 Tabla de frecuencias para {col}:")
    print(df1[col].value_counts())  # Cuenta las ocurrencias de cada categoría
    print("-" * 50)
````

La variable cualitativa más frecuente es **Neighborhood**

````python
#Analisis de correlación

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar solo las variables cuantitativas
cuantitativas = df1.select_dtypes(include=['int64', 'float64'])

# Calcular la matriz de correlación
corr_matrix = cuantitativas.corr()

# Visualizar la matriz de correlación con un heatmap
plt.figure(figsize=(18, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, cbar_kws={"shrink": 0.8}, annot_kws={"size": 8})
plt.title('Matriz de Correlación')
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=10)
plt.show()
````

Existen varias variables con una correlación alta, por ejemplo el tamaño del sótano **TotalBsmtSF** y el primer piso **1stFlrSF**, estan estrechamente relacionados. Asi como el mañano del garage con la capacidad de autos, lo que sugiere que una de ellas podría eliminarse en un modelo para evitar multicolinealidad. No se encontraron correlaciones negativas fuertes, lo que indica que la mayoría de las relaciones son positivas.

````python
#Analisis de tendencias temporales

# Agrupar por año de construcción y calcular la media de una variable 
tendencia_anual = df1.groupby('YearBuilt')['MoSold'].mean()

# Graficar la tendencia temporal
plt.figure(figsize=(10, 6))
tendencia_anual.plot(kind='line', marker='o', color='b')
plt.title('Tendencia del Mes de venta por año de construcción')
plt.xlabel('Año de Construcción')
plt.ylabel('Mes de venta promedio')
plt.grid(True)
plt.show()
````

````python
# Agrupar por año de construcción y calcular la media de una variable 
tendencia_anual = df1.groupby('YearBuilt')['YrSold'].mean()

# Graficar la tendencia temporal
plt.figure(figsize=(10, 6))
tendencia_anual.plot(kind='line', marker='o', color='b')
plt.title('Tendencia del año de venta por año de construcción')
plt.xlabel('Año de Construcción')
plt.ylabel('año de venta promedio')
plt.grid(True)
plt.show()
````
